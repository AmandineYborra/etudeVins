```python
prod2015 = production2015.copy()
prod2016 = production2016.copy()
prod2017 = production2017.copy()
```

```python
prod2015['Total Blanc'] = prod2015.apply( lambda row : (row["Qt AOP Blanc (hl)"] + row["Qt IGP Blanc (hl)"]+ row["Qt VSIG Blanc (hl)"]), axis = 1 ) 
prod2015['Total Rouge'] = prod2015.apply( lambda row : (row["Qt AOP Rouge (hl)"] + row["Qt IGP Rouge (hl)"]+ row["Qt VSIG Rouge (hl)"]), axis = 1 ) 
prod2015['Total Rosé'] = prod2015.apply( lambda row : (row["Qt AOP Rosé (hl)"] + row["Qt IGP Rosé (hl)"]+ row["Qt VSIG Rosé (hl)"]), axis = 1 ) 
prod2015['Total Prod'] = prod2015.apply( lambda row : (row["Prod commercialisable (hl)"] + row["Prod non commercialisable (hl)"]), axis = 1 ) 
prod2015.head(1)
```

```python
prod2016['Total Blanc'] = prod2016.apply( lambda row : (row["Qt AOP Blanc (hl)"] + row["Qt IGP Blanc (hl)"]+ row["Qt VSIG Blanc (hl)"]), axis = 1 ) 
prod2016['Total Rouge'] = prod2016.apply( lambda row : (row["Qt AOP Rouge (hl)"] + row["Qt IGP Rouge (hl)"]+ row["Qt VSIG Rouge (hl)"]), axis = 1 ) 
prod2016['Total Rosé'] = prod2016.apply( lambda row : (row["Qt AOP Rosé (hl)"] + row["Qt IGP Rosé (hl)"]+ row["Qt VSIG Rosé (hl)"]), axis = 1 ) 
prod2016['Total Prod'] = prod2016.apply( lambda row : (row["Prod commercialisable (hl)"] + row["Prod non commercialisable (hl)"]), axis = 1 ) 
prod2016.head(1)
```

```python
prod2017['Total Blanc'] = prod2017.apply( lambda row : (row["Qt AOP Blanc (hl)"] + row["Qt IGP Blanc (hl)"]+ row["Qt VSIG Blanc (hl)"]), axis = 1 ) 
prod2017['Total Rouge'] = prod2017.apply( lambda row : (row["Qt AOP Rouge (hl)"] + row["Qt IGP Rouge (hl)"]+ row["Qt VSIG Rouge (hl)"]), axis = 1 ) 
prod2017['Total Rosé'] = prod2017.apply( lambda row : (row["Qt AOP Rosé (hl)"] + row["Qt IGP Rosé (hl)"]+ row["Qt VSIG Rosé (hl)"]), axis = 1 ) 
prod2017['Total Prod'] = prod2017.apply( lambda row : (row["Prod commercialisable (hl)"] + row["Autres production (hl)"]), axis = 1 ) 
prod2017.head(1)
```

```python
prod2015 = prod2015.loc[:, ['Sup totale vignes (ha)', 'Qt Cognac ou Armagniac (hl)', 'Total Blanc', 'Total Rouge', 'Total Rosé', 'Total Prod']]
prod2015.head()
```

```python
prod2016 = prod2016.loc[:, ['Sup totale vignes (ha)', 'Qt Cognac ou Armagniac (hl)', 'Total Blanc', 'Total Rouge', 'Total Rosé', 'Total Prod']]
prod2016.head()
```

```python
prod2017 = prod2017.loc[:, ['Sup totale vignes (ha)', 'Qt Cognac ou Armagniac (hl)', 'Total Blanc', 'Total Rouge', 'Total Rosé', 'Total Prod']]
prod2017.head()
```
